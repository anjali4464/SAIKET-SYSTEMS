```python
import pandas as pd 
df=pd.read_csv('updated_dataset_after_task_1.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
# Calculate mean of 'TotalCharges' excluding NaN values
mean_total_charges = df['TotalCharges'].mean()

# Fill NaN values in 'TotalCharges' with its mean
df['TotalCharges'].fillna(mean_total_charges, inplace=True)
```

```python
bool_cols = df.select_dtypes(include=['bool']).columns
bool_cols
```

```python
#converting all bool values to integer 
for col in bool_cols:
    df[col]=df[col].apply(lambda x: 1 if x == True else 0)
```

```python
x=df.drop(columns=['Churn_Yes', 'Churn_No', 'customerID'])
y = df['Churn_Yes'].apply(lambda x: 1 if x == True else 0)
```

Identify key factors influencing churn and Propose data-driven strategies to reduce churn.suggest actions.


```python



# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Build RandomForest model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Get feature importances
feature_importances = model.feature_importances_
features = x.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances}).sort_values(by='Importance', ascending=False)
print(importance_df)
```

Calculate the lifetime value (LTV) of customers.

```python
# Calculate ARPU (using MonthlyCharges as an estimate)
#Average Revenue per Customer (ARPU)
arpu = df['TotalCharges'].mean()

# Calculate Average Lifespan (using tenure)
average_lifespan = df['tenure'].mean()

# Calculate LTV(life time value)
#Lifetime Value or LTV is an estimate of the average revenue that a customer will generate throughout their lifespan as a customer.
# we can call ltv as clv(customer lifetime value)
ltv = arpu * average_lifespan
```

```python
print("Lifetime Value or LTV: ",ltv)
```
